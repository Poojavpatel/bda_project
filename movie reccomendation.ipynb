```python
movie_matrix = df.pivot_table(index='userId')
movie_matrix.head()
```

```python
movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
movie_matrix.head()
```

```python
movie_matrix.describe()
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
# Let’s assume that a user has watched Air Force One (1997) and Contact (1997).
ratings.loc[['Air Force One (1997)','Contact (1997)']]
```

```python
movie_matrix.head()
```

```python
AFO_user_rating = movie_matrix['Air Force One (1997)']
AFO_user_rating.head()
```

```python
contact_user_rating = movie_matrix['Contact (1997)']
contact_user_rating.head()
```

```python
# Corrwith computes the pairwise correlation of rows or columns of two dataframe objects
# get the correlation between each movie's rating and the ratings of the Air Force One movie.

similar_to_air_force_one=movie_matrix.corrwith(AFO_user_rating)
similar_to_air_force_one.head()
```

```python
similar_to_air_force_one.sort_values(ascending=False).head(5)
```

```python
similar_to_contact = movie_matrix.corrwith(contact_user_rating)
similar_to_contact.head()
```

```python
similar_to_contact.sort_values(ascending=False).head(5)
```

```python
corr_AFO = pd.DataFrame(similar_to_air_force_one, columns=['correlation'])
corr_AFO.dropna(inplace=True)
corr_AFO.head()
```

```python
corr_contact = pd.DataFrame(similar_to_contact, columns=['Correlation'])
corr_contact.dropna(inplace=True)
corr_contact.head()
```

```python
# Some of the movies have very few ratings. We can fix this by setting a threshold for the number of ratings. 
# From the histogram earlier we saw a sharp decline in number of ratings from 100. We shall therefore set this as the threshold
```

```python
corr_AFO = corr_AFO.join(ratings['number_of_ratings'])
corr_AFO .head()
```

```python
corr_contact = corr_contact.join(ratings['number_of_ratings'])
corr_contact.head()
```

```python
corr_AFO[corr_AFO['number_of_ratings'] > 50].sort_values(by='correlation', ascending=False).head(10)
```

```python
corr_contact[corr_contact['number_of_ratings'] > 50].sort_values(by='Correlation', ascending=False).head(10)
```

```python
# Hence we recommend Tombstone (1993) with correlation 0.823566
# and Wolf of Wall Street(2013) with correlation 0.875113
```
